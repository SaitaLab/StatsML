### 全ての画像で右下の14*14ピクセルを選択する方法

```python
my_slice = train_images[:,14:,14:]
plt.imshow(my_slice[4],cmap=plt.cm.binary)
plt.show()
```

### 負のインデックスを指定することで、終端を基準とする相対位置を表現可能
画像の中央から14*14ピクセルをくりぬく方法は

```python
my_slice = train_images[:,7:-7,7:-7]
plt.imshow(my_slice[4],cmap=plt.cm.binary)
plt.show()
```

## データバッチ
データテンソルの最初の軸をサンプル軸（サンプル次元）と呼ぶ。
（MNISTでは数字の画像がサンプルである）
ディープラーニングの過程でデータセット全体は一度に処理されずバッチという単位に分割する。
（MNISTではバッチサイズは128である）
MNISTのバッチは以下のように指定する。

```python
# 1つめのバッチから先頭の画像を取り出す
batch = train_images[:128]
plt.imshow(batch[0],cmap=plt.cm.binary)
plt.show()
```

```python
# 2つめのバッチから先頭の画像を取り出す
batch = train_images[128:256]
plt.imshow(batch[0],cmap=plt.cm.binary)
plt.show()
```

```python
# n番目のバッチから先頭の画像を取り出す
n=3
batch = train_images[128*n:128*(n+1)]
plt.imshow(batch[0],cmap=plt.cm.binary)
plt.show()
```
